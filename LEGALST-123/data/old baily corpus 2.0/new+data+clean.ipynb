```python
import lxml.etree as et
import pandas as pd

# LOAD XML AND XSLT
doc = et.parse("test.xml")
xsl = et.parse("test3.xsl")

# RUN TRANSFORMATION
transformer = et.XSLT(xsl)
result = transformer(doc)

print(result)

# BUILD DICTIONARY OF VALUES FROM FLATTER RESULT
data = {}
for i in result.xpath('/*/*'):    
    data[i.tag] = i.text

# BIND TO DATAFRAME
trial_df = pd.DataFrame([data])

print(trial_df)
```
